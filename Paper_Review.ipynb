## 1. Introduction

In this paper, the author proposed a new generative model estimation procedure in order to solve the main problem of generative models, which is the difficulty of approximating many intractable probabilistic computations that arise in maximum likelihood estimation and related strategies, and the difficulty of leveraging the benefits of piece-wise linear units in the generative context[1]. So that generative models may have more impact.
The Generative Adversarial Nets proposed by the author basically have two parts, a discriminative model that learns to determine whether a sample is from the model distribution or the data distribution, which means, to tell whether the data it receive is a “real” data; and a generative model that produces “fake” data with a random noise and try to use it without detected by the discriminative model. The training procedure is done when the generative model can generate almost real data that can’t be distinguished by the discriminative model.
In the Generative Adversarial Nets, we can train both two models using only the backpropagation, as well as the dropout algorithms [2] to avoid over-fitting. Approximate inference or Markov chains are not needed in this generative net, so that the computational cost is greatly reduced.



## 2. Content

### > Related Works

In the first section of this paper numbers of related works are reviewed . First of all,  author presented the main problem of undirected graphical models with latent variables(eg.RBMs and DBMs), which quantity as well as gradient estimation are mostly rely on the Markov chain Monte Carlo methods, while mixing poses can cause much trouble for this method. 
Also, DBNs, representing hybrid models which involve a single undirected layer and several directed layers, always encounter computational difficulties .
There are also other criteria like score matching and NCE to avoid the computational problem, but these criteria need learned probability density analyzed into a normalization constant. However in numbers of generative models, we can’t even acquire a tractable unnormalized probability density.
At last, the adversarial nets have a better performance of backpropagation comparing to techniques without defining a probability distribution explicitly eg. GSN framework, while it may have problems with unbounded activation when used in a feedback loop.

### > Explaining Adversarial Nets

In this section, the author uses figures and some simple formulas to explain what is adversarial nets doing and how it is accomplishing the mission that train the generative model as well as updating the discriminative model.

### > Theoretical Results

In this section, the author presents a theoretical analysis of adversarial nets, analyzing this method in a mathematical way how the training procedure allows one to recover the data generating distribution as the generator and the discriminator are given enough capacity. 
Using mathematical formulas to proof that this adversarial game between the generative model and the discriminative model has a global optimum for pg = pdata, which means that the distribution of generated data and real data is completely the same, thus achieving the desired result.


### > Experiments

In this section, the author trained the adversarial nets using a wide range of data including MNIST,TFD and CIFAR-10. Then the author used Parzen window-based log-likelihood estimates and evaluate the performance of adversarial nets by comparing it to other generative models. Then in this paper, the author also tried to visualize samples from the model. As a result, we can see that although it is not performing better than all those existing methods, it’s performance has convinced us that it is at least a very powerful competitor. Besides, as the adversarial nets have just be introduced, it still have much potential and can be considered as one of the most promising methods for unsupervised learning in complex distribution in recent years. 

### > Advantages and Disadvantages

In this section, the author talks about both advantages and disadvantages of this newly introduced framework.
One of the disadvantage is that we have to make sure that the training procedure of generator G and discriminator D are trained at the same pace, otherwise it is very likely to encounter a “the Helvetica scenario”, which means that it will have a very pool ability dealing with diversity data and will finally leads to a mode collapses.
There are also advantages of this framework, one of it is that there is no need for a Markov chain, so that it is not computationally costly, it will have a lower training difficulty with a higher training efficiency and it can contain numbers of different functions. Also, statistically, it has the advantage that updated directly with data examples, but with the gradient, which will increase diversity of generated samples.Besides, the adversarial nets can finally generate very sharp distributions, which means that the samples generated by this framework will be more easier for people to understand.



### > Conclusion

In this section, the author talked about the overall situation of the adversarial nets, and its possible future extensions.
1.Firstly the generation procedure of GAN doesn’t require complicated operations such as sampling sequence, which means that it can adopt a new data C and start to sample and infer directly.
2.Learned approximate inference can be performed by training an auxiliary net . With the advantage of it , the inference net may be trained for a fixed generator net after the generator net has finished training.
3.One can approximate much wider conditionals using a family of conditional models that share parameters.
4. Also this framework can be further used in semi-supervised learning, because the training procedure of generative adversarial nets doesn’t require data to be labeled. 
5.Besides, its training can be faster if the method for G and D is improved or it has a better sample distribution.



## 3. Innovation

This paper has presented us the generative adversarial model, which is a new generative model using a new method. It is presented to solve the major problem of current existing generative models, for example, computationally complicated ,low efficiency and low accuracy in most cases. 
GAN, which has been presented by this paper, is definitely have a great impact to the development of generative models. As a generative method, GAN effectively solves the problem of generating data that can be interpreted naturally. Especially in the case of generating high-dimensional data, the neural network that has adopted by this model does not restrict the dimension of generation, which greatly broadens the scope of generating data samples that it can deal with. Besides, the structure of the model can integrate all kinds of loss functions and increase the degree of freedom of design. Also, GAN training process innovatively takes the adversarial of two models as training criteria and can be trained by only using the highly successful back propagation and dropout algorithms. The training process of the model does not require a Markov chain, which is computationally costly with a lower efficiency, nor does it need any unrolled approximate inference network, all these improvements will greatly reduce the training difficulty and enhance the training efficiency of generative models.
What’s more, the generation of GAN does not need complicated sampling sequence. Which means that it can adopt sample and infer new samples directly, which improves the productivity of new samples. The adversarial model training method not being updated directly with data examples, but with the gradient that flows through the discriminator, thus increasing the diversity of generated samples. It doesn’t require a label, which has made it suitable for further usage in semi-supervised learning. At last, the model generated are mostly sharp distributions, which means that the samples generated by this framework will be more easier for people to understand.
